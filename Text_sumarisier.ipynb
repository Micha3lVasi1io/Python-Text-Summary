# Load Summarisation Pipeline

```python
from transformers import pipeline

# Create a pre-trained summarisation pipeline
summariser = pipeline('summarization')
```

# Summarise Text

```python
# Get the user input
text = input("Enter the text you want to summarize: ")

# MAX_LENGTH = Maximum number of words to summarise back to user.
# MIN_LENGTH = Minimum number of words to return back to user.
# DO_SAMPLE (false) = Return back each word, with highest probability of making sense.
summary = summariser(text, max_length=130, min_length=30, do_sample=False)

# Display the summarised text, only taking the text from a dictionary.
summary_text = summary[0]['summary_text']
print("\nSummary:")
print(summary_text)
```
